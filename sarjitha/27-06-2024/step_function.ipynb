```python
# import the necessary libraries
import numpy as np 
import matplotlib.pyplot as plt

```

```python
x=np.linspace(-10,10,100)
```

```python
def stepFunction(z):
    return np.where (z>=0,1,0)
y_step=stepFunction(x)
plt.plot (x,y_step,label='step Function',color='b')
plt.title ('step Function')
plt.xlabel('input')
plt.ylabel('output')
plt.grid(True)
plt.show()
          
```

```python
def sigmoid(z):
    return 1 /(1+np.exp(-z))
y_sigmoid=sigmoid(x)
plt.plot(x,y_sigmoid,label='sigmoid',color='b')
plt.title('sigmiod')
plt.ylabel('output')
plt.grid(True)
plt.show()
```

```python
def tanh(z):
    return np.tanh(z)
y_tanh=tanh(x)
plt.plot(x,y_tanh,label='tanh',color='r')
plt.title('tanh')
plt.ylabel('output')
plt.grid(True)
plt.legend()
plt.show()
```

```python

def relu(z):
    return np.maximum(0, z)

y_relu = relu(x)


plt.plot(x, y_relu, label='ReLU', color='g')
plt.title('ReLU')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define Leaky Relu Function
def leaky_relu(z, alpha=0.01):
    return np.where(z > 0, z, alpha * z)

y_leaky_relu = leaky_relu(x)

plt.plot(x, y_leaky_relu, label='Leaky ReLU', color='y')
plt.title('Leaky ReLU')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def softmax(z):
    exp_z = np.exp(z)
    return exp_z / np.sum(exp_z, axis=0)
y_softmax = softmax(x)

plt.plot(x, y_softmax, label='Softmax', color='m')
plt.title('Softmax')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```
